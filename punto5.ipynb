```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Coordenadas polares
theta = np.linspace(0, 2*np.pi, 150)
r = np.linspace(0, 1, 100)
T, R = np.meshgrid(theta, r)
X = R*np.cos(T)
Y = R*np.sin(T)
Z = np.sqrt(X**2 + Y**2)

fig = plt.figure(figsize=(12, 4))

ax = fig.add_subplot(1, 3, 1, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Coordenadas polares')

# Coordenadas rectangulares
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)

ax = fig.add_subplot(1, 3, 2, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Coordenadas rectangulares')

# Coordenadas cilndricas
theta = np.linspace(0, 2*np.pi, 100)
z = np.linspace(-1, 1, 100)
T, Z = np.meshgrid(theta, z)
X = np.cos(T)
Y = np.sin(T)
R = np.sqrt(X**2 + Y**2)

ax = fig.add_subplot(1, 3, 3, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Coordenadas cilndricas')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Funcin objetivo
def f(x):
    return x**2

# Derivada de la funcin objetivo
def df(x):
    return 2*x

# Parmetros del gradiente descendente
x_start = 5
learning_rate = 0.1
n_iterations = 20

# Gradiente descendente
x = x_start
x_values = [x]
for i in range(n_iterations):
    x -= learning_rate * df(x)
    x_values.append(x)

# Grfico de la funcin objetivo y los valores de x obtenidos con el gradiente descendente
x_range = np.linspace(-6, 6, 100)
plt.plot(x_range, f(x_range))
plt.scatter(x_values, f(np.array(x_values)), c='r')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Funcin objetivo
def f(x):
    return x[0]**2 + x[1]**2

# Gradiente de la funcin objetivo
def grad_f(x):
    return 2 * x

# Parmetros del gradiente descendente
alpha = 0.1 # Tasa de aprendizaje
tol = 1e-6 # Tolerancia para el criterio de parada
max_iter = 1000 # Nmero mximo de iteraciones

# Inicializacin del vector de parmetros
x = np.array([2.0, 2.0])

# Almacenamiento de la trayectoria del vector de parmetros
x_path = [x]

# Bucle del gradiente descendente
for i in range(max_iter):
    # Actualizacin del vector de parmetros
    x = x - alpha * grad_f(x)
    x_path.append(x)
    
    # Criterio de parada
    if np.linalg.norm(grad_f(x)) < tol:
        break

# Conversin de la trayectoria a un array de NumPy
x_path = np.array(x_path)

# Grfico de la funcin objetivo y la trayectoria del vector de parmetros
x1 = np.linspace(-2.5, 2.5, 100)
x2 = np.linspace(-2.5, 2.5, 100)
X1, X2 = np.meshgrid(x1, x2)
Y = X1**2 + X2**2

plt.contour(X1, X2, Y, levels=np.logspace(-0.5, 1.5, 10))
plt.plot(x_path[:,0], x_path[:,1], '-o', color='red')
plt.show()
```
